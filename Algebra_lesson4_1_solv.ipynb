__Теорема__ 

Необходимым и достаточным условием совместности системы из $m$ уравнений с $n$ неизвестными является равенство между собой рангов матрицы коэффициентов $A$ и расширенной матрицы $\tilde A$

$$rank A=rank \tilde A.$$

Причем:<br>
1) если $rankA=rank \tilde A=n$, где $n$ — число неизвестных, то система определена, т. е. имеет единственное решение;<br>
2) если $rankA=rank \tilde A<n$, то система имеет бесконечное количество решений;<br>
3) если $rankA<rank \tilde A$, то система несовместна.

```python
def check_rang_matrix(matrix_A, vector_b):
    matrix_A_ext = np.column_stack([A, b])
    
    rank_A = np.linalg.matrix_rank(matrix_A)
    rank_A_ext = np.linalg.matrix_rank(matrix_A_ext)
    n = len(matrix_A[0])
    
    result = 'Система несовместна, решений нет'
    if rank_A == rank_A_ext:
        if rank_A == n:
            result = 'Система определена и имеет единственное решение'
        else:
            result = 'Система совместна и имеет бесконечное кол-во решений'

    print('Ранг матрицы равен:\t%d\nРанг расширенной матрицы равен:\t%d\nЧисло неизвестных:\t%d\n%s'%(
        rank_A,
        rank_A_ext,
        n,
        result
    ))
```

```python
A = np.array([[3, -1, 1], [2, -5, -3], [1, 1, -1]])
b = np.array([4, -17, 0])

check_rang_matrix(A, b)
```

```python
A = np.array([[2, -4, 6], [1, -2, 3], [3, -6, 9]])
b = np.array([1, -2, 5])

check_rang_matrix(A, b)
```

```python
A = np.array([[1, 2, 5], [3, 1, -8]])
b = np.array([4, -2])

check_rang_matrix(A, b)
```
